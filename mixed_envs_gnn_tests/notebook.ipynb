```python
import torch
from torch_geometric.data import Data
from torch_geometric.loader import DataLoader

n = 10
f = 2
x = torch.rand((n,f))
edge_index = (torch.rand((2,n*n))*n).type(torch.long)
data1 = Data(x=x, edge_index=edge_index)
print(data1)
print(data1.x)
print(data1.edge_index)

x = torch.rand((n,f))
edge_index = (torch.rand((2,n*n))*n).type(torch.long)
data2 = Data(x=x, edge_index=edge_index)
print(data2)
print(data2.x)
print(data2.edge_index)


dl = DataLoader([data1,data2],batch_size=2)
print(dl)
data_batched = next(iter(dl))
print(data_batched)
print(data_batched.x)
print(data_batched.edge_index)
print(data_batched.batch)

```

```python
from torch_geometric.nn.models import MLP


mlp = MLP(in_channels=4, hidden_channels=10,out_channels=3, num_layers=3,norm="LayerNorm")
a = torch.rand((5,4))
batch = torch.tensor([0,0,1,1,1])
print(batch)
print(a)
b = mlp(a,batch=batch)
print(b)
a[0,2] = 10
# print(a)
bb = mlp(a,batch=batch)
print(bb)
bb = mlp(a,batch=batch)
print(bb)
```
